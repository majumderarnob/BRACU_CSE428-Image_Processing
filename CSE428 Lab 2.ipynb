```python
import numpy as np
from skimage import exposure, io, color, img_as_float
import matplotlib.pyplot as plt

np.random.seed(1000)
```

```python
from google.colab import drive
drive.mount("/content/gdrive")
```

```python
I = np.random.rand(3,3)
print(I)
io.imshow(I,cmap='Greys_r') #imshow is use for showing image
```

```python
# Addition
I_add = I + 0.5
print(I_add)
io.imshow(I_add,cmap='Greys_r')
```

```python
# I_add[I_add>1.0] = 1.0

I_add = np.clip(I_add, 0, 1)

print(I_add)
io.imshow(I_add)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(I,cmap='Greys_r')
plt.title("original")

plt.subplot(1,2,2)
io.imshow(I_add,cmap='Greys_r')
plt.title("Addition by 0.5")
plt.show()
```

```python
#Multipicaltion
I_multiplication = I * 0.5
print(I_multiplication)
io.imshow(I_multiplication,cmap='Greys_r')
```

```python
# I_multiplication[I_multiplication>1.0] = 1.0

I_multiplication = np.clip(I_multiplication, 0, 1)

print(I_multiplication)
io.imshow(I_multiplication)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(I_multiplication,cmap='Greys_r')
plt.title("original")

plt.subplot(1,2,2)
io.imshow(I_multiplication,cmap='Greys_r')
plt.title("Multiplying by 0.5")
plt.show()
```

```python
I_sat = img_as_float(io.imread("/content/gdrive/MyDrive/CSE428 Lab/landsat_down.jpeg"))
print(type(I_sat))
#plt.figure(figsize=(2,2))

io.imshow(I_sat)
```

```python
print(f"Max Value = {np.max(I_sat)}")
print(f"Min Value = {np.min(I_sat)}")
```

```python
hist, bins_hist = exposure.histogram(I_sat, nbins=256, normalize=True)
print(np.where(hist > 1e-3))
plt.plot(bins_hist, hist)
```

```python
I_sat_stretched = np.clip(4*(I_sat-0.1), 0, 1)
io.imshow(I_sat_stretched)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
io.imshow(I_sat)
plt.title("original")

plt.subplot(1,2,2)
io.imshow(I_sat_stretched)
plt.title("After Contrast Stretching")
plt.show()
```

```python
hist_orig, bins_hist = exposure.histogram(I_sat, nbins=256, normalize=True)
hist_stretched, bins_hist = exposure.histogram(I_sat_stretched, nbins=256, normalize=True)

plt.figure()
plt.subplot(1,2,1)
plt.plot(bins_hist, hist_orig)
plt.title("Original")

plt.subplot(1,2,2)
plt.plot(bins_hist, hist_stretched)
plt.title("After Contrast Stretching")

plt.show()
```
